```python
#importando a biblioteca
import pandas as pd
```

```python
#Leitura de arquivos
df1 = pd.read_excel("Aracaju.xlsx")
df2 = pd.read_excel("Fortaleza.xlsx")
df3 = pd.read_excel("Natal.xlsx")
df4 = pd.read_excel("Recife.xlsx")
df5 = pd.read_excel("Salvador.xlsx")
```

```python
#leituda dos dados 
df1.head()
```

```python
#leituda dos dados 
df2.head()
```

```python
#leituda dos dados 
df3.head()
```

```python
#leituda dos dados 
df4.head()
```

```python
#leituda dos dados 
df5.head()
```

```python
#juntando todos os arquivos
df = pd.concat([df1,df2,df3,df4,df5])
```

```python
#fazer uma amostra da planilha
df.sample(5)
```

```python
#verificando o tipo de cada coluna
df.dtypes
```

```python
#alterando o tipo de dado de cada lojaID
df["LojaID"] = df["LojaID"].astype("object")
```

```python
df.dtypes
```

```python
df.head()
```

Valores Faltantes

```python
#Consultando linhas com valores faltantes
df.isnull().sum()
```

```python
#substituindo os valores nulos pela média
df["Vendas"].fillna(df["Vendas"].mean(), inplace=True)
```

```python
df.isnull().sum()
```

```python
df["Vendas"].mean()
```

```python
df.sample(15)
```

```python
#Substituir falores faltantes por Zero
df["Vendas"].fillna(0, inplace=True)
```

```python
#Apagando as linhas com valores nulos
df.dropna(inplace=True)
```

```python
#apagando as linhas com valores nulos com base apenas em uma coluna
df.dropna(subset=["Vendas"], inplace=True)
```

```python
#Removendo linhas que estejam com valores faltantes em todas as colunas
df.dropna(how="all",inplace=True)
```

Criando colunas novas

```python
#criando a coluna de receita
df["Receita"] = df["Vendas"].mul(df["Qtde"])
```

```python
df.head()
```

```python
#Cria uma coluca com a quantidade de produtos
df["Receita/Vendas"] = df["Receita"] / df["Vendas"]
```

```python
df.head()
```

```python
#Retorna a maior receita
df["Receita"].max()
```

```python
#retornando a menor receita
df["Receita"].min()
```

```python
#retornando as maiores receitas linha completa
df.nlargest(3, "Receita")
```

```python
#retornando as menores receitas linha completa
df.nsmallest(3, "Receita")
```

```python
#Agrupando por cidade
df.groupby("Cidade")["Receita"].sum()
```

```python
#Ordenando o conjunto de dados
df.sort_values("Receita", ascending=False).head(10)
```

Trabalhando com datas


```python
#transformando a coluna de data em tipo inteiro
df["Data"] = df["Data"].astype("int64")
```

```python
#verificando o tipo de cada coluna
df.dtypes
```

```python
#Transformando coluna de data em data
df["Data"] = pd.to_datetime(df["Data"])
```

```python
df.dtypes
```

```python
#agrupando por ano
df.groupby(df["Data"].dt.year)["Receita"].sum()
```

```python
#Criano uma nova coluna com ano 
df["Ano_Venda"] = df["Data"].dt.year
```

```python
df.sample(5)
```

```python
#Extraindo o mes e o dia
df["mes_venda"],df["dia_venda"] = (df["Data"].dt.month, df["Data"].dt.day)
```

```python
df.sample(5)
```

```python
#Retornando a data mais antiga
df["Data"].min()
```

```python
#Calculando a diferença de dias
df["diferença_dias"] = df["Data"] - df["Data"].min()
```

```python
df.sample(5)
```

```python
#criando a coluna de trimestre
df["Trimestre_venda"] = df["Data"].dt.quarter
```

```python
df.sample(5)
```

```python
#filtrando as vendas de 2019 do mês de março
vendas_marco_19 = df.loc[(df["Data"].dt.year == 2019) & (df["Data"].dt.month == 3)]
```

```python
vendas_marco_19.sample(20)
```
